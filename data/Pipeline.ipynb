## Pipeline para tratamento das imagens de treinamento do modelo

```python
import os
import numpy as np
import cv2 as ocv
import pandas as pd
```

Função responsável por traduzir imagens em matrizes.

```python
def img_to_df(path: str, label: int):
    length = 0
    rows = None

    for file in os.listdir(path):
        img = ocv.imread(f"{path}\\{file}")
        # img = ocv.resize(img, None, fx=0.3, fy=0.3)
        img = ocv.cvtColor(img, ocv.COLOR_BGR2GRAY)
        
        reshaped = np.reshape(img, (1, img.shape[0] * img.shape[1]))

        if length == 0:
            rows = reshaped
        else:
            rows = np.concatenate((rows, reshaped))
        
        length += 1

    class_list = [label] * length

    return pd.concat([pd.DataFrame(rows), pd.DataFrame(class_list, columns=["Jumping"])], axis=1)
```

Transformação do arquivos das pastas de imagens fontes para um arquivo csv de treinamento.

```python
jumping = img_to_df(f"{os.getcwd()}\\jumping", 1)
not_jumping = img_to_df(f"{os.getcwd()}\\no-jumping", 0)
```

```python
csv = pd.concat([jumping, not_jumping])
csv
```

```python
csv.to_csv("training_data.csv", index=False)
```
