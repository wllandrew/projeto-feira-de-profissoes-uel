## Pipeline para tratamento das imagens de treinamento do modelo

```python
import os
import numpy as np
import cv2 as ocv
import pandas as pd
```

```python
SIZE_JUMP = len([f for f in os.listdir(f"{os.getcwd()}\\raw\\jumping")])
SIZE_NOTJUMP = len([f for f in os.listdir(f"{os.getcwd()}\\raw\\no-jumping")])
```

```python
print(SIZE_JUMP)
print(SIZE_NOTJUMP)
```

```python
global rows

for p in range(SIZE_JUMP):
    img = ocv.imread(f"{os.getcwd()}\\raw\\jumping\\image_{p}.png")
    img = ocv.resize(img, None, fx=0.3, fy=0.3)
    img = ocv.cvtColor(img, ocv.COLOR_BGR2GRAY)
    
    x = img.shape
    reshaped = np.reshape(img, (1,x[0] * x[1]))

    if p == 0:
        rows = reshaped
    else:
        rows = np.concatenate((rows, reshaped))
```

```python
for p in range(SIZE_NOTJUMP):
    img = ocv.imread(f"{os.getcwd()}\\raw\\no-jumping\\image_{p}.png")
    img = ocv.resize(img, None, fx=0.3, fy=0.3)
    img = ocv.cvtColor(img, ocv.COLOR_BGR2GRAY)
    
    x = img.shape
    reshaped = np.reshape(img, (1, x[0] * x[1]))

    rows = np.concatenate((rows, reshaped))
```

```python
csv = pd.DataFrame(rows)
csv
```

```python
rotulos = ["Jumping"] * SIZE_JUMP + ["Not Jumping"] * SIZE_NOTJUMP 
csv = pd.concat([csv, pd.DataFrame(rotulos, columns=["Class"])], axis=1)
csv
```

```python
csv.to_csv("ans.csv")
```
